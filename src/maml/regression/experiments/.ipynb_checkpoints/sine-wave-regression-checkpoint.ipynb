```python
import os
import sys
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
from copy import deepcopy

import torch
import torch.nn as nn
from torch.optim.optimizer import Optimizer

# configure module & syspath
module_path = os.path.abspath(os.path.join('../../../../'))
if module_path not in sys.path:
    sys.path.append(module_path)

from src.maml.regression.datasets.sine_wave import SineWave
from src.maml.regression.models.mlp import MLP
```

```python
device = 'cpu'
dev = torch.device(device)
plt.rcParams['figure.figsize'] = (10, 4)
```

```python
@torch.no_grad()
def generate_task() -> SineWave:
    """
    Samples amplitude and phase, then returns a Sine Wave function corresponding to it.

    Returns:
        Returns a neural network that is tuned for few-shot learning.
    """
    amp = torch.rand(1).item() * 4.9 + 0.1
    phase = torch.rand(1).item() * np.pi

    return SineWave(amp, phase)
```

```python
def MAML(
  num_training_epochs: int,
  num_tasks: int,
  sample_size_k: int,
  learning_rate: float,
  device: str
):
    """
    Runs MAML on in the supervised regression setting as described in the MAML paper.

    Args:
        num_training_epochs (int): The number of epochs to run MAML over.
        num_tasks (int): Number of tasks to be sampled from the distribution.
        sample_size_k (int): The number of samples to be used for each of the tasks.
        learning_rate (float): Learning rate to be used for the optimizers.
        device (str): Device to run MAML on (can either be set to CPU or GPU)

    Returns:
        Returns a neural network that is tuned for few-shot learning.
    """
    
    """
    @todo check if the MAML algorithm can be further broken down, how much of this can be reused for classificaiton / reinforcement learning.
    @todo check if the learning rates for internal and meta updates should be different.
    """

    meta_model = MLP(input_dim=1, hidden_dim=40, output_dim=1).to(device)
    meta_optim = torch.optim.Adam(meta_model.parameters(), lr=learning_rate)

    training_loss = list()
    inner_training_loops = 1

    for _ in tqdm(range(num_training_epochs)):
        theta_prime = dict()
        data_prime = dict()
        
        """
        Sample tasks from the distribution.
        """
        regression_tasks = [generate_task() for _ in range(num_tasks)]

        """
        Iterate over each task in order to:
            - Evaluate with respect to K examples
            - Compute adapted parameters with gradient descent.
        """
        for i, task in enumerate(regression_tasks):
            meta_model_copy = MLP(input_dim = 1, hidden_dim = 40, output_dim = 1).to(device)
            meta_model_copy.load_state_dict(meta_model.state_dict())

            local_optim = torch.optim.SGD(meta_model_copy.to(device).parameters(), lr=learning_rate)

            for _ in range(inner_training_loops):
                """
                Sample k points from a task, evaluate loss with respect to K examples.
                """
                x_batch, y_batch = task.sample(sample_size_k)
                loss_function = nn.MSELoss()
                loss = loss_function(meta_model_copy(x_batch.to(device)), y_batch.to(device))

                """
                Compute adapted parameters iwth gradient descent.
                """ 
                local_optim.zero_grad()
                loss.backward()
                local_optim.step()
                continue

            """
            For the meta-update, we need to:
                - Track adapted parameter "theta prime".
                - Sample additional data from the current task that will be used to check out-of-sample performance.
            """
            theta_prime[i] = meta_model_copy.state_dict()

            # @todo check if this update makes a large difference.
            x_prime_batch, y_prime_batch = task.sample(25)
            data_prime[i] = (x_prime_batch, y_prime_batch)
            pass

        """
        While making the meta-update:
            - Sample data points for each task.
            - Compute loss for each of the "theta prime" parameters.
            - Aggregate gradients for losses corresponding to each "theta prime".
        """
        meta_optim.zero_grad()
        meta_grads = [torch.zeros_like(p) for p in meta_model.parameters()]
        batch_training_loss = list()

        for i, _ in enumerate(regression_tasks):
            model_theta_prime = MLP(input_dim = 1, hidden_dim = 40, output_dim = 1).to(device)
            model_theta_prime.load_state_dict(theta_prime[i])

            """
            Compute out-of-sample loss for each of the tasks.
            """
            x_prime_batch, y_prime_batch = data_prime[i]

            loss_function = nn.MSELoss()
            loss = loss_function(model_theta_prime(x_batch.to(device)), y_batch.to(device))
            loss.backward()
            
            batch_training_loss.append(loss.item())

            """
            Then, aggregate the gradients.
            """
            for grad, param in zip(meta_grads, model_theta_prime.parameters()):
                grad += param.grad

        """
        Update parameter updates using aggregated gradients.
        """
        for param, grad in zip(meta_model.parameters(), meta_grads):
            param.grad = grad
            meta_optim.step()
            pass
        
        training_loss.append(np.mean(batch_training_loss))

    return meta_model
```

```python
meta_model = MAML(num_training_epochs=70000, sample_size_k=10, learning_rate=1e-3, device=device, num_tasks=10)
```

```python
# code for K-shot training
def k_shot_tuning(model: nn.Module, task: SineWave, batch_size_k: int, gradient_steps: int, learning_rate: float, device: str):
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
    x_batch, y_batch = task.sample(batch_size_k)

    for epoch in range(gradient_steps + 1):
        loss_fct = nn.MSELoss()
        loss = loss_fct(model(x_batch.to(device)), y_batch.to(device))
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        pass

    return model
```

```python
sine_wave_task = generate_task()
ground_truth_x = torch.linspace(-5, 5, 50)
ground_truth_y = sine_wave_task.amplitude * torch.sin(ground_truth_x + sine_wave_task.phase)

# pre-update
predicted_y = meta_model(ground_truth_x[..., None])

# plot the results after finetuning the model.
plt.title('Pre-update')
plt.plot(ground_truth_x.data.numpy(), ground_truth_y.data.numpy(), c='red', label='Ground Truth')
plt.plot(ground_truth_x.data.numpy(), predicted_y.data.numpy(), c='orange', label='Predicted Values')
plt.legend()
plt.show()
```

```python
# K-shot training with specified number of gradient steps.
tuned_meta_model = k_shot_tuning(
    model=deepcopy(meta_model), 
    task=sine_wave_task, 
    batch_size_k=5, 
    gradient_steps=1,
    learning_rate=1e-2,
    device=device
)

predicted_y = tuned_meta_model(ground_truth_x[..., None])

plt.title('K=5')
plt.plot(ground_truth_x.data.numpy(), ground_truth_y.data.numpy(), c='red', label='Ground Truth')
plt.plot(ground_truth_x.data.numpy(), predicted_y.data.numpy(), c='orange', label='Predicted Values')
plt.legend()
plt.show()
```

```python
# K-shot training with specified number of gradient steps.
tuned_meta_model = k_shot_tuning(
    model=deepcopy(meta_model), 
    task=sine_wave_task, 
    batch_size_k=10, 
    gradient_steps=10, 
    learning_rate=1e-2,
    device=device
)

predicted_y = tuned_meta_model(ground_truth_x[..., None])

plt.title('K=5')
plt.plot(ground_truth_x.data.numpy(), ground_truth_y.data.numpy(), c='red', label='Ground Truth')
plt.plot(ground_truth_x.data.numpy(), predicted_y.data.numpy(), c='orange', label='Predicted Values')
plt.legend()
plt.show()
```
